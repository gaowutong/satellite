```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import datetime
```

```python
errs = pd.read_csv('errors.csv', index_col=0)
errs = errs.transpose()
errs.index = [datetime.datetime.strptime(dts, "%Y-%m-%d %H:%M:%S") for dts in errs.index]
errs[errs > 1e4] = np.nan
# errs = errs.iloc[:, -3:]
# errs.to_csv('errs_cleaned.csv')
errs
```

```python
import matplotlib.dates as mdates
```

```python
ax = errs.plot(figsize=(15, 10), use_index=True)
fmt = mdates.DateFormatter("%d %H:%M")
ax.get_xaxis().set_major_formatter(fmt)
ax.set_xlabel('Datetime (dd HH:MM)', fontsize=16)
ax.set_ylabel('Errors (m)', fontsize=16)
ax.set_title('G01 Ephemeris Calculation Errors of different interval in 2018-08-24', fontsize=18)
plt.savefig('errors.jpg')
```
